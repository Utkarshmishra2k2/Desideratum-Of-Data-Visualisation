Relative Stacked Bar Plots

```python
Data03.head()        
```

```python
Data03["Total"]=Data03.sum(axis=True)    
Data03                
```

```python
#Here I Adding Fraction Of All Cases  
Data03['ConfirmedFraction']=Data03['Confirmed']/Data03['Total']     
Data03['RecoveredFraction']=Data03['Recovered']/Data03['Total']      
Data03['DeceasedFraction']=Data03['Deceased']/Data03['Total']         

Data03             
```

```python
# Here I Ploted The Fraction Of Each Section
fig=plt.gcf();                                                                 
fig.set_size_inches(30,9)                                                       
plt.bar(Data03.index,Data03.ConfirmedFraction,color="Orange");                                       
plt.bar(Data03.index,Data03.RecoveredFraction,bottom=Data03.ConfirmedFraction,color="Green");                 
plt.bar(Data03.index,Data03.DeceasedFraction,bottom=Data03.ConfirmedFraction+Data03.RecoveredFraction,color="Red"); 
plt.xticks(rotation=90);                                                        
```

```python
Data03=Data03.sort_values('ConfirmedFraction',ascending=0)      
```

```python
# Here I Ploted The Fraction In  Dscending Order
fig=plt.gcf();                                                                  
fig.set_size_inches(30,9)                                                     
plt.bar(Data03.index,Data03.ConfirmedFraction,color="Orange");                                       
plt.bar(Data03.index,Data03.RecoveredFraction,bottom=Data03.ConfirmedFraction,color="Green");                 
plt.bar(Data03.index,Data03.DeceasedFraction,bottom=Data03.ConfirmedFraction+Data03.RecoveredFraction,color="Red"); 
plt.xticks(rotation=90);                                                       
```

```python
Data03=Data03.sort_values('DeceasedFraction',ascending=1)      
```

```python
# Here I Ploted The Fraction In Ascending Order
fig=plt.gcf();                                                                 
fig.set_size_inches(30,9)                                                     
plt.bar(Data03.index,Data03.ConfirmedFraction,color="Orange");                                       
plt.bar(Data03.index,Data03.RecoveredFraction,bottom=Data03.ConfirmedFraction,color="Green");                 
plt.bar(Data03.index,Data03.DeceasedFraction,bottom=Data03.ConfirmedFraction+Data03.RecoveredFraction,color="Red"); 
plt.xticks(rotation=90);                                                       
```

```python
# Here The Added The Sum Values Of All Graph
Data03=Data03.sort_values('Total',ascending=1)
fig=plt.gcf();                                                                  
fig.set_size_inches(30,9)                                                       
plt.bar(Data03.index,Data03.Confirmed,color="Orange");                                       
plt.bar(Data03.index,Data03.Recovered,bottom=Data03.Confirmed,color="Green");                 
plt.bar(Data03.index,Data03.Deceased,bottom=Data03.Confirmed+Data03.Recovered,color="Red"); 
plt.xticks(rotation=90);                                                      

for U,values in enumerate(Data03.index):                                         
    y=Data03.loc[values,"Total"]+100                                                  
    x=U
    plt.text(x,y,str(y),ha="center");                               
```

```python
# Here I Added The The Horizontal Bar Graph
Data03=Data03.sort_values('Total',ascending=1)
fig=plt.gcf();                                                                  
fig.set_size_inches(30,9)                                                     
plt.barh(Data03.index,Data03.Confirmed,color="Orange");                                       
plt.barh(Data03.index,Data03.Recovered,left=Data03.Confirmed,color="Green");                 
plt.barh(Data03.index,Data03.Deceased,left=Data03.Confirmed+Data03.Recovered,color="Red"); 
plt.xticks(rotation=90);                                                        
```
