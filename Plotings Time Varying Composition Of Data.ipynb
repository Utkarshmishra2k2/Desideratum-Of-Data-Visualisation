```python
Data=Data['states_daily']     
```

```python
da=pd.read_json('Data04.json') 
```

```python
Data04=pd.json_normalize(Data)     
```

```python
Data05= Data04.tail(3)                          
Data05.drop("date",axis=1,inplace=True)         
Data05.drop("dateymd",axis=1,inplace=True)      
Data05.set_index("status",inplace=True)         
Data05=Data05.T                                 
Data05=Data05.apply(pd.to_numeric)             
```

```python
Data05    
```

**Pivot**

Pivot Means Reshaping Data Frame

```python
Data04         
```

```python
Data05=Data04[["up","date","status"]]       
```

```python
Data05["up"]=pd.to_numeric(Data05["up"])    
Data05["date"]=pd.to_datetime(Data05["date"])   
Data05                     
```

**Piovt Is Very Important.**

**Here Pivot Starts .**


Here If You All Status Is In Single Column, Which Make It Difficult To Operate.

To Tackle This I Haave To Convert THis In 03 Different Rows i.e Confirmed,Recovered,Deceased.

Here The Structure Of New Data Set Is Date|Confirmed|Recovered|Deceased.

```python
Data05=Data05.pivot_table(values="up",columns="status",index="date")   
```

```python
fig=plt.gcf();                                                                 
fig.set_size_inches(15,5)                                                       
plt.stackplot(Data05.index,Data05.Confirmed,Data05.Recovered,Data05.Deceased,      
              labels=["Confirmed","Recovered","Deceased"],                          
              colors=["green","orange","red"]);                                    
plt.xticks(rotation=90);                       
plt.legend();                                 

```

```python
#Stacked Bar Plot Of Fraction Of Data
fig=plt.gcf();                                                               
fig.set_size_inches(15,5)                                                     
plt.stackplot(Data05.index,Data05.Confirmed/Data05.sum(axis=1),Data05.Recovered/Data05.sum(axis=1),Data05.Deceased/Data05.sum(axis=1),      
              labels=["Confirmed","Recovered","Deceased"],                         
              colors=["green","orange","red"]);                                  
plt.xticks(rotation=90);                        
plt.legend();                                   
```

Plotting Stacked Bar Area's Fraction Data For Any State

```python
def StackedArea(State):                         
    Data05=Data04[[State,"date","status"]]      
    Data05[State]=pd.to_numeric(Data05[State])      
    Data05["date"]=pd.to_datetime(Data05["date"])   
    Data05=Data05.pivot_table(values=State,columns="status",index="date")  
    #Stacked Bar Plot Of Fraction Of Data
    fig=plt.gcf();                                                                 
    fig.set_size_inches(15,5)                                                      
    plt.stackplot(Data05.index,Data05.Confirmed/Data05.sum(axis=1),Data05.Recovered/Data05.sum(axis=1),Data05.Deceased/Data05.sum(axis=1),    
              labels=["Confirmed","Recovered","Deceased"],                         
              colors=["green","orange","red"]);                                     
    plt.xticks(rotation=90);                        
    plt.legend();                                    
```
